2.2
Classification of Personal Data and Sensitive Data and Results

```python
#Employee data classification

original_employees_df_columns = ['Id. de empleado','Apellido','Nombre','Cargo','Direcci√≥n','Tratamiento', 'Fecha de nacimiento', 'Fecha de contrataci√≥n', 'Direcci√≥n', 'Ciudad', 'Sigla', 'C√≥digo postal', 'Pa√≠s', 'Tel√©fono de domicilio', 'Extensi√≥n', 'Foto', 'Notas', 'Jefe']
employee_personal_data = ['Id. de empleado', 'Apellido', 'Nombre', 'Fecha de nacimiento', 'Direcci√≥n', 'Tel√©fono de domicilio','Extensi√≥n']
employee_sensitive_data = ['Foto', 'Notas']

def verify_personal_data_present_employees(original_columns, personal_data):
     for col in personal_data:
         if col in original_columns.columns:
             return True
     else:
         return False
    
def verify_sensitive_data_present_employees(original_columns, sensitive_data):
     for col in sensitive_data:
         if col in original_columns.columns:
             return True
     else:
         return False

employees_df_to_pandas['Datos personales'] = verify_personal_data_present_employees(employees_df_to_pandas.loc[:,original_employees_df_columns], employee_personal_data)
employees_df_to_pandas['Datos sensibles'] = verify_sensitive_data_present_employees(employees_df_to_pandas.loc[:,original_employees_df_columns], employee_sensitive_data)
```

```python
#Client data classification

columns_df_clients_original = ['Id. de cliente', 'Nombre de compa√±√≠a', 'Nombre del contacto', 'Cargo del contacto', 'Direcci√≥n', 'Ciudad', 'Regi√≥n', 'C√≥digo postal', 'Pa√≠s', 'Tel√©fono', 'Fax']
client_personal_data = ['Id. de cliente', 'Nombre del contacto', 'Direcci√≥n', 'Tel√©fono', 'Fax']
client_sensitive_data = []

def verify_personal_data_present_clients(original_columns, personal_data):
     for col in personal_data:
         if col in original_columns.columns:
             return True
     else:
         return False
    
def verify_sensitive_data_present_clients(original_columns, sensitive_data):
     for col in sensitive_data:
         if col in original_columns.columns:
             return True
     else:
         return False

clients_df_to_pandas['Datos personales'] = verify_personal_data_present_clients(clients_df_to_pandas.loc[:,columns_df_clients_original], client_personal_data)
clients_df_to_pandas['Datos sensibles'] = verify_sensitive_data_present_clients(clients_df_to_pandas.loc[:,columns_df_clients_original], client_sensitive_data)
```

3.3 Implementation of Rules and Reporting

NOTE: Descriptions of the general operation of the rules, as well as criticality, dimension, affected columns, and rationale are found in the report (Section 3.2). The rule index number in the report is detailed in parentheses in the comments for each quality rule below.

```python
# Rule 1 (3.2.1)

def get_uniqueness_client_id(client_id):
     return ~client_id.duplicated(keep=False)

df1 = pd.DataFrame(get_uniqueness_client_id(clients_df_to_pandas.loc[:, 'client ID']).value_counts())

df1.rename(columns={'client ID': 'REGLA 3.2.1'}, inplace=True)
df1
```

```python
#Rule 2 (3.2.2)

def get_completeness_client_id(client_id):
     return client_id is not None and client_id != ''

df2 = pd.DataFrame(get_completeness_client_id(clients_df_to_pandas.loc[:, 'client ID']).value_counts())

df2.rename(columns={'client ID': 'REGLA 3.2.2'}, inplace=True)
df2
```

```python
# Rule 3 (3.2.3)

def get_customer_id_consistency(entered_customer_id_value, contrasted_df_column):
     return entered_customer_id_value in contrasted_df_column.values

df3 = pd.DataFrame(orders_df_to_pandas['Customer'].apply(get_customer_id_consistency, contrasted_df_column=clients_df_to_pandas['Customer ID']).value_counts())

df3.rename(columns={'Customer': 'REGLA 3.2.3'}, inplace=True)
df3
```

```python
#Rule 4 (3.2.4)

def get_order_id_consistency(entered_order_id_value, contrasted_df_column):
     return entered_order_id_value in contrasted_df_column.values

df4 = pd.DataFrame(order_details_df_to_pandas['OrderId'].apply(get_order_id_consistency, contrasted_df_column=orders_df_to_pandas['OrderId']).value_counts())

df4.rename(columns={'order ID': 'REGLA 3.2.4'}, inplace=True)
df4
```

```python
#Rule 5 (3.2.5)

def get_precision_customer_id(customer_id):
     pattern = '^[A-Z]{5}$'
     try:
         return customer_id.str.contains(pattern, regex=True)
     except ValueError:
         return False

df5 = pd.DataFrame(get_precision_customer_id(clients_df_to_pandas['Customer ID']).value_counts())

df5.rename(columns={'client ID': 'REGLA 3.2.5'}, inplace=True)
df5
```

```python
#Rule 6 (3.2.6)

def get_order_quantity_validity(entered_quantity):
     return entered_quantity is not None and (float(entered_quantity) > 0)
    
df6 = pd.DataFrame(order_details_df_to_pandas['Quantity'].apply(get_order_quantity_validity).value_counts())

df6.rename(columns={'Amount': 'REGLA 3.2.6'}, inplace=True)
df6
```

```python
#Rule 7 (3.2.7)

def get_price_validity_per_order_unit(price_entered):
     try:
         price = float(price_entered.replace('$', '').replace(',', ''))
         return price > 0
     except (ValueError, TypeError):
         return False
    
df7 = pd.DataFrame(order_details_df_to_pandas['Price per unit'].apply(get_price_validity_per_order_unit(price_entered).value_counts()))

df7.rename(columns={'Unit Price': 'REGLA 3.2.7'}, inplace=True)
df7
```

```python
#Rule 8 (3.2.8)

def get_opportunity_order_date(entered_date):
     try:
         converted_date = pd.to_datetime(entered_date)
         return converted_date >= pd.to_datetime("1995-05-25")
     except ValueError:
         return False

df8 = pd.DataFrame(orders_df_to_pandas['Order Date'].apply(get_opportunity_order_date).value_counts())

df8.rename(columns={'Order date': 'REGLA 3.2.8'}, inplace=True)
df8

    
```

```python
#Rule 9 (3.2.9)

def get_product_image_validity(input_file):
     return input_file is not None and input_file.endswith(".BMP")

df9 = pd.DataFrame(categories_df_to_pandas['Image'].apply(get_product_image_validity).value_counts())

df9.rename(columns={'Image': 'REGLA 3.2.9'}, inplace=True)
df9
```

```python
# Rule 10 (3.2.10)

def get_validity_product_category(entered_category):
    return entered_category is not None

df10 = pd.DataFrame(categories_df_to_pandas['Category name'].apply(get_validity_product_category).value_counts())

df10.rename(columns={'Category name': 'REGLA 3.2.10'}, inplace=True)
df10
```
